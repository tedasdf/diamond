### Polynomial Time Scaling

```python
T = 1

a0 = 0
a1 = 0
a2 = 3 / (T**2)
a3 = -2 / (T**3)

# Define the functions
s = lambda t: a0 + a1*t + a2*t**2 + a3*t**3
s_dot = lambda t: a1 + 2*a2*t + 3*a3*t**2
s_ddot = lambda t: 2*a2 + 6*a3*t

# Time values
t_vals = np.linspace(0, T, 100)
s_vals = s(t_vals)
s_dot_vals = s_dot(t_vals)
s_ddot_vals = s_ddot(t_vals)

# Create 3 side-by-side subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot s(t)
axes[0].plot(t_vals, s_vals, color='blue')
axes[0].set_title("s(t)")
axes[0].set_xlabel("t")
axes[0].set_ylabel("Value")
axes[0].grid(True)

# Plot s'(t)
axes[1].plot(t_vals, s_dot_vals, color='green')
axes[1].set_title("s'(t)")
axes[1].set_xlabel("t")
axes[1].grid(True)

# Plot s''(t)
axes[2].plot(t_vals, s_ddot_vals, color='red')
axes[2].set_title("s''(t)")
axes[2].set_xlabel("t")
axes[2].grid(True)

plt.tight_layout()
plt.show()
```
